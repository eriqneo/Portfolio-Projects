```python
from bs4 import BeautifulSoup
import requests
import time
import datetime
import csv
```

```python
URL = 'https://www.jumia.co.ke/catalog/?q=shoes'
headers = {"User-Agent": "Mozilla/5.0 (X11; Ubuntu; Linux x86_64; rv:121.0) Gecko/20100101 Firefox/121.0"}

page = requests.get(URL, headers=headers)

if page.status_code == 200:
    soup1 = BeautifulSoup(page.content, 'html.parser')
```

```python
# Find all product containers on the page
product_containers = soup1.find_all(class_='core')

    # Get today's date
today_date = datetime.date.today().strftime('%Y-%m-%d')

    # CSV file path
csv_filename = f'jumia_products_{today_date}.csv'

with open(csv_filename, 'w', newline='', encoding='utf-8') as csvfile:
        fieldnames = ['Name', 'Price', 'Price Before', 'Ratings', 'Date']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)

        # Write header to the CSV file
        writer.writeheader()

        # Iterate through each product container
        for container in product_containers:
            # Extract information for each product
            name = container.find(class_='name').get_text(strip=True)
            price = container.find(class_='prc').get_text(strip=True)
            price_before = container.find(class_='old').get_text(strip=True)
            ratings = container.find(class_='stars _s').get_text(strip=True)

            # Write information to the CSV file
            writer.writerow({'Name': name, 'Price': price, 'Price Before': price_before, 'Ratings': ratings, 'Date': today_date})

            # Print or process the information for each product
            print("Name:", name)
            print("Price:", price)
            print("Price Before:", price_before)
            print("Ratings:", ratings)
            print("Date:", today_date)
            print("---")

            print(f"CSV file '{csv_filename}' has been created.")

        else:
             print(f"Failed to retrieve the page. Status code: {page.status_code}")
```
