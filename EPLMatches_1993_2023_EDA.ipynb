```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt

import seaborn as sns
```

```python
epl = pd.read_csv('/content/drive/MyDrive/epl.csv')

epl.head()
```

```python
epl.shape
```

creating a points column

```python
#with FTR Column, create points column, where H AND A =3 AND D = 1

epl['Points'] = epl['FTR'].map({'H': 3, 'A': 3, 'D': 1})

epl.head()
```

```python
#Making another column for Goal Difference, later we shall see if we can make a cumulative one for each team

epl['GD'] = epl['HomeGoals'] - epl['AwayGoals']

epl.head()
```

```python
#how many games were played each season

epl['Season_End_Year'].value_counts()
```

```python
#how many games were played each season in totality

epl['Season_End_Year'].value_counts().sum()
```

more games were played in '93, '94, 95. Let us find the number of games played by a team

```python
#Lets find out how many game a team had to play back then. As from 1996 teams played 38 games, you can check with the below formula

# Filter the DataFrame for the years 1993, 1994, and 1995
years_of_interest = [1993] #[1996]
epl_filtered = epl[epl['Season_End_Year'].isin(years_of_interest)]

# Concatenate the 'Home' and 'Away' columns to create a Series containing all the teams
teams_home = epl_filtered['Home']
teams_away = epl_filtered['Away']
all_teams = pd.concat([teams_home, teams_away])

# Count the occurrences of each team in the concatenated Series
team_game_counts = all_teams.value_counts()

# Display the number of games played by each team
team_game_counts.head()
```

wh more games?

```python
#more games only means one thing: more teams, see below; As from 1996 teams were reduced to 20

# Filter the DataFrame for the year 1993
epl_1993 = epl[epl['Season_End_Year'] == 1993]

# Extract unique teams from both 'Home' and 'Away' columns
unique_teams_home = epl_1993['Home'].unique()
unique_teams_away = epl_1993['Away'].unique()

# Combine unique teams from both columns
all_unique_teams = set(unique_teams_home) | set(unique_teams_away)

# Count the number of unique teams
num_teams_1993 = len(all_unique_teams)

print(f"Number of teams that played in 1993: {num_teams_1993}")
```

```python
#Teams with highest home wins

top_home_results = epl.pivot_table('Points', 'Home', aggfunc = 'sum').sort_values('Points', ascending=False).head()

bottom_home_results = epl.pivot_table('Points', 'Home', aggfunc = 'sum').sort_values('Points', ascending=False).tail()

epl_home_results = pd.concat([top_home_results, bottom_home_results])

epl_home_results

#Manutd and Totenham, have the highest home wins
#Blaxkpool and Barnsley have the lowest home wins
```
